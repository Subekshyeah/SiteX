```python
import pandas as pd
import folium
from folium.plugins import HeatMap
import os
```

```python
base_path = '../backend/Data/CSV_Reference/final/'
files = ['banks_final.csv', 'cafe_final.csv', 'education_final.csv', 'health_final.csv', 'other_final.csv', 'temples_final.csv']
dfs = []
for file in files:
    df = pd.read_csv(os.path.join(base_path, file))
    dfs.append(df)
```

```python
combined_df = pd.concat(dfs, ignore_index=True)
combined_df = combined_df.dropna(subset=['lat', 'lng'])
poi_data = combined_df[['lat', 'lng']].values.tolist()
```

```python
m = folium.Map(location=[combined_df['lat'].mean(), combined_df['lng'].mean()], zoom_start=10)
HeatMap(poi_data).add_to(m)
m
```
